```python
# 필요 라이브러리 불러오기
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
```

```python
# 이미지 불러오고 리사이즈 부터 정규화하는 함수
def img_normalize(path, height, width):
    image_files = os.listdir(path) # 파일 경로에 있는 폴더안에 모든 파일을 리스트 형태로 저장
    images = []
    # 폴더안에 있는 모든 파일을 하나씩 for 문으로 보기
    for file in image_files:
        img = cv2.imread(os.path.join(path, file)) # 파일경로 + 이미지 이름 결합(join)해서 하나의 디렉토리로 만든걸 읽기
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Blue Gree Red 에서 RGB 로 변환
        images.append(img) # images 리스트에 RGB 로 변환된 이미지를 추가

    # 리사이징
    resized_images = [cv2.resize(img, (height, width)) for img in images]

    # 이미지 정규화 
    normalized_images = [img / 255.0 for img in resized_images]
    return normalized_images

# 파일 경로
img_path = r'C:\Users\hkpar\Downloads\archive\Pistachio_Image_Dataset\Pistachio_Image_Dataset\Kirmizi_Pistachio'

# 파라미터 가로 세로 설정
h, w = (100, 100)
img_normalized = img_normalize(img_path, h, w)

len(img_normalized)
```
